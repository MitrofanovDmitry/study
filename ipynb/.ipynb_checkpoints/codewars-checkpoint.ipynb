```python
# https://www.codewars.com/kata/52e1476c8147a7547a000811/train/python

from re import search
regex="^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)[A-Za-z\d]{6,}$"

assert bool(search(regex, 'fjd3IR9')) == True
assert bool(search(regex, 'ghdfj32')) == False
assert bool(search(regex, 'DSJKHD23')) == False
assert bool(search(regex, 'dsF43')) == False
assert bool(search(regex, '4fdg5Fj3')) == True
assert bool(search(regex, 'DHSJdhjsU')) == False
assert bool(search(regex, 'fjd3IR9.;')) == False
```

```python
# https://www.codewars.com/kata/564057bc348c7200bd0000ff/train/python

def thirt(n):
    while True:
        a = list(reversed([int(x) for x in str(n)]))
        b = [v*((10**i)%13) for (i,v) in enumerate(a)]
        p = n; n = sum(b);
        if n == p: break
    return n

assert thirt(1234567) == 87
assert thirt(8529) == 79
assert thirt(85299258) == 31
assert thirt(5634) == 57
assert thirt(1111111111) == 71
assert thirt(987654321) == 30
```

```python
# https://www.codewars.com/kata/5626b561280a42ecc50000d1/train/python

def dig_pow(n):
    return sum(int(x)**y for y,x in enumerate(str(n), 1))

def sum_dig_pow(a, b): 
    return [x for x in range(a,b + 1) if x == dig_pow(x)]

assert sum_dig_pow(1, 10) == [1, 2, 3, 4, 5, 6, 7, 8, 9]
assert sum_dig_pow(1, 100) == [1, 2, 3, 4, 5, 6, 7, 8, 9, 89]
assert sum_dig_pow(10, 89) ==  [89]
assert sum_dig_pow(10, 100) ==  [89]
assert sum_dig_pow(90, 100) == []
assert sum_dig_pow(89, 135) == [89, 135]
```

```python
# https://www.codewars.com/kata/5626b561280a42ecc50000d1/train/python

def dig_pow(num):
    s = 0
    n = num
    c = len(str(num))
    while n > 0:
        s += (n % 10) ** c
        n //= 10; c -= 1;
    return s

def sum_dig_pow(a, b):
    result = []
    for num in range(a, b + 1):
        if num == dig_pow(num):
            result.append(num)
    return sorted(result)

assert sum_dig_pow(1, 10) == [1, 2, 3, 4, 5, 6, 7, 8, 9]
assert sum_dig_pow(1, 100) == [1, 2, 3, 4, 5, 6, 7, 8, 9, 89]
assert sum_dig_pow(10, 89) ==  [89]
assert sum_dig_pow(10, 100) ==  [89]
assert sum_dig_pow(90, 100) == []
assert sum_dig_pow(89, 135) == [89, 135]
```

```python
# https://www.codewars.com/kata/554a44516729e4d80b000012/train/python

def nbMonths(startPriceOld, startPriceNew, savingperMonth, percentLossByMonth):

    months,savings = 0,0
    
    while startPriceOld + savings < startPriceNew:
        months += 1
        savings += savingperMonth
        if months % 2 == 0: percentLossByMonth += 0.5
        startPriceOld -= startPriceOld * percentLossByMonth/100
        startPriceNew -= startPriceNew * percentLossByMonth/100
    
    return [months, round(startPriceOld + savings - startPriceNew)]

assert nbMonths(2000, 8000, 1000, 1.5) == [6, 766]
assert nbMonths(12000, 8000, 1000, 1.5) == [0, 4000]
```

```python
# https://www.codewars.com/kata/54a91a4883a7de5d7800009c/train/python

def increment_string(s):
    
    count = 0
    
    for c in s[::-1]:
        if not c.isdigit(): break
        count += 1
            
    num = (int(s[-count:]) if count else 0) + 1
    return s[0:len(s)-count] + str(num).zfill(count)

assert increment_string('foo') == 'foo1'
assert increment_string('foobar001') == 'foobar002'
assert increment_string('foobar1') == 'foobar2'
assert increment_string('foobar00') == 'foobar01'
assert increment_string('foobar99') == 'foobar100'
assert increment_string('foobar099') == 'foobar100'
assert increment_string('') == '1'
assert increment_string('5227)tQ),|3(u5375483O90043')
```

```python
# https://www.codewars.com/kata/52c31f8e6605bcc646000082/train/python

def two_sum(numbers, target):
    for i in range(len(numbers)-1):
        for j in range(i+1,len(numbers)):
            if numbers[i] + numbers[j] == target:
                return (i,j)

assert sorted(two_sum([1,2,3], 4)) == [0,2]
assert sorted(two_sum([1234,5678,9012], 14690)) == [1,2]
assert sorted(two_sum([2,2,3], 4)) == [0,1]
```

```python
# https://www.codewars.com/kata/554ca54ffa7d91b236000023/train/python

def delete_nth(order,max_e):
    r = []
    for o in order:
        if r.count(o) < max_e: r.append(o)
    return r

assert delete_nth([20,37,20,21], 1) == [20,37,21]
assert delete_nth([1,1,3,3,7,2,2,2,2], 3) == [1, 1, 3, 3, 7, 2, 2, 2]
```

```python
# https://www.codewars.com/kata/554ca54ffa7d91b236000023/train/python

def delete_nth(order,max_e):
    counts = dict()
    result = []
    
    for item in order:
        if item not in counts: 
            counts[item] = 0
        
        counts[item] += 1
        
        if counts[item] <= max_e: 
            result.append(item)
            
    return result

assert delete_nth([20,37,20,21], 1) == [20,37,21]
assert delete_nth([1,1,3,3,7,2,2,2,2], 3) == [1, 1, 3, 3, 7, 2, 2, 2]
```

```python
# https://www.codewars.com/kata/59437bd7d8c9438fb5000004

import re

def alphabet_war(battlefield):
    
    if '#' not in battlefield:
        return battlefield.replace('[','').replace(']','')
    
    # Here (?=...) is a lookahead assertion
    matches = re.findall(r'(?=(?:^|\])([a-z#]*\[[a-z]*\][a-z#]*)(?:\[|$))', battlefield)
    survived = [re.findall(r'\[([a-z]*)\]', m)[0] for m in matches if m.count('#') < 2]
    return ''.join(survived)

assert alphabet_war('[a]#[b]#[c]') == 'ac'
assert alphabet_war('[a]#b#[c][d]') == 'd'
assert alphabet_war('[a][b][c]') == 'abc'
assert alphabet_war('##a[a]b[c]#') =='c'
assert alphabet_war('abde[fgh]ijk') == 'abdefghijk'
assert alphabet_war('ab#de[fgh]ijk') == 'fgh'
assert alphabet_war('ab#de[fgh]ij#k') == ''
assert alphabet_war('##abde[fgh]ijk') == ''
assert alphabet_war('##abde[fgh]') == ''
assert alphabet_war('##abcde[fgh]') == ''
assert alphabet_war('abcde[fgh]') == 'abcdefgh'
assert alphabet_war('##abde[fgh]ijk[mn]op') == 'mn'
assert alphabet_war('#abde[fgh]i#jk[mn]op') == 'mn'
assert alphabet_war('[ab]adfd[dd]##[abe]dedf[ijk]d#d[h]#') == 'abijk'
```
