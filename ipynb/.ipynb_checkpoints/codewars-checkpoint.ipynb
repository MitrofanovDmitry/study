```python
# https://www.codewars.com/kata/55b7bb74a0256d4467000070/train/python

def proper_fractions(n):
    phi = n > 1 and n
    for p in range(2, int(n ** .5) + 1):
        if not n % p:
            phi -= phi // p
            while not n % p:
                n //= p
    if n > 1: phi -= phi // n
    return phi

assert proper_fractions(1) == 0
assert proper_fractions(2) == 1
assert proper_fractions(5) == 4
assert proper_fractions(15) == 8
assert proper_fractions(25) == 20
```

```python
# https://www.codewars.com/kata/556c04c72ee1147ff20000c9/train/python

def valid(a):
    
    pairs = dict()
    players = set(''.join([i for o in a for i in o]))
    
    for day in a:
        if len(day) != len(a[0]): return False
        
        for player in players:
            if ''.join(day).count(player) != 1: return False
        
        for group in day:
            if len(group) != len(a[0][0]): return False
            
            for i in range(len(group)-1):
                for j in range(i+1,len(group)):
                    k = tuple(sorted((group[i],group[j])))
                    if not pairs.get(k): pairs[k] = 0
                    pairs[k] += 1
                    if pairs[k] > 1: return False
    
    return True

s = [ ["AB"] ]
assert valid(s) == True, "Two player, one day works"

s = [
    ["AB", "CD"],
    ["AD", "BC"],
    ["BD", "AC"]]
assert valid(s) == True, "Four players, three days works"

s = [
    ['ABCD', 'EFGH', 'IJKL', 'MNOP', 'QRST'],
    ['AEIM', 'BJOQ', 'CHNT', 'DGLS', 'FKPR'],
    ['AGKO', 'BIPT', 'CFMS', 'DHJR', 'ELNQ'],
    ['AHLP', 'BKNS', 'CEOR', 'DFIQ', 'GJMT'],
    ['AFJN', 'BLMR', 'CGPQ', 'DEKT', 'HIOS']]
assert valid(s) == True, "Wolfram MathWorld case works"

s = [
    ["AB", "CD", "EF", "GH"],
    ["AC", "BD", "EG", "FH"],
    ["AD", "CE"],
    ["AE", "BG", "CH", "FD"]]
assert valid(s) == False, "Must reject day with fewer groups"

s = [
    ["ABC", "DEF"],
    ["ADE", "CBF"]]
assert valid(s) == False, "Must reject B & C playing together twice"
```

```python
# https://www.codewars.com/kata/54bb6f887e5a80180900046b/train/python

def longest_palindrome (s):
    max_len = 1 if len(s) else 0
    for i in range(len(s)-1):
        for j in range(i+2,len(s)+1):
            if s[i:j] == s[i:j][::-1]:
                if j - i > max_len:
                    max_len = j - i
    return max_len

assert longest_palindrome("a") == 1
assert longest_palindrome("aa") == 2
assert longest_palindrome("baa") == 2
assert longest_palindrome("aab") == 2
assert longest_palindrome("abcdefghba") == 1
assert longest_palindrome("baablkj12345432133d") == 9
```

```python
# https://www.codewars.com/kata/554a44516729e4d80b000012/train/python

def nbMonths(startPriceOld, startPriceNew, savingperMonth, percentLossByMonth):

    months,savings = 0,0
    
    while startPriceOld + savings < startPriceNew:
        months += 1
        savings += savingperMonth
        if months % 2 == 0: percentLossByMonth += 0.5
        startPriceOld -= startPriceOld * percentLossByMonth/100
        startPriceNew -= startPriceNew * percentLossByMonth/100
    
    return [months, round(startPriceOld + savings - startPriceNew)]

assert nbMonths(2000, 8000, 1000, 1.5) == [6, 766]
assert nbMonths(12000, 8000, 1000, 1.5) == [0, 4000]
```

```python
# IQ Test

def iq_test(numbers):
    even,odd = [],[]
    numbers = map(int, numbers.split())
    for i,n in enumerate(numbers):
        if n % 2 == 0: even.append(i+1)
        else: odd.append(i+1)
    if len(even) == 1: return even[0]
    if len(odd) == 1: return odd[0]
    return None

assert iq_test("2 4 7 8 10") == 3
assert iq_test("1 2 2") == 1
```

```python
# https://www.codewars.com/kata/57814d79a56c88e3e0000786

def decrypt(text, n):
    
    if not text or n < 0: return text
    
    for i in range(n):
        l = len(text)
        a = [' '] * l
        a[1::2] = text[0:l//2]
        a[0::2] = text[l//2:l]
        text = ''.join(a)
    
    return text

def encrypt(text, n):
    
    if not text or n < 0: return text
    
    for i in range(n):
        text = ''.join(text[1::2] + text[0::2])
    
    return text

assert encrypt("This is a test!", 0) == "This is a test!"
assert encrypt("This is a test!", 1) == "hsi  etTi sats!"
assert encrypt("This is a test!", 2) == "s eT ashi tist!"
assert encrypt("This is a test!", 3) == " Tah itse sits!"
assert encrypt("This is a test!", 4) == "This is a test!"
assert encrypt("This is a test!", -1) == "This is a test!"
assert encrypt("This kata is very interesting!", 1) == "hskt svr neetn!Ti aai eyitrsig"

assert decrypt("This is a test!", 0) == "This is a test!"
assert decrypt("hsi  etTi sats!", 1) == "This is a test!"
assert decrypt("s eT ashi tist!", 2) == "This is a test!"
assert decrypt(" Tah itse sits!", 3) == "This is a test!"
assert decrypt("This is a test!", 4) == "This is a test!"
assert decrypt("This is a test!", -1) == "This is a test!"
assert decrypt("hskt svr neetn!Ti aai eyitrsig", 1) == "This kata is very interesting!"

assert encrypt("", 0) == ""
assert decrypt("", 0) == ""
assert encrypt(None, 0) == None
assert decrypt(None, 0) == None
```

```python
# https://www.codewars.com/kata/555615a77ebc7c2c8a0000b8/train/python

def sell(given, price, at_hand):

    aside, change = [], []
    at_hand.append(given)
    at_hand.sort()
    
    while len(at_hand):
        bill = at_hand.pop()
        fits = sum(change) + bill <= (given - price)
        (change if fits else aside).append(bill)
    
    at_hand.extend(aside)
    
    if sum(change) != (given - price):
        at_hand.extend(change)
        at_hand.remove(given)
        return False
    
    return True

def tickets(people):
    
    price = 25
    at_hand = []
    
    for given in people:
        if sell(given, price, at_hand) == False: 
            return 'NO'
    
    return 'YES'

assert tickets([25, 25, 50]) == "YES"
assert tickets([25, 100]) == "NO"
```

```python
# https://www.codewars.com/kata/563cf89eb4747c5fb100001b/train/python

def remove_smallest(a):
    b = a.copy()
    if len(b): b.remove(min(b))
    return b

assert remove_smallest([1, 2, 3, 4, 5]) == [2, 3, 4, 5]
assert remove_smallest([5, 3, 2, 1, 4]) == [5, 3, 2, 4]
assert remove_smallest([1, 2, 3, 1, 1]) == [2, 3, 1, 1]
assert remove_smallest([]) == []
```

```python
# https://www.codewars.com/kata/563b662a59afc2b5120000c6/train/python

def nb_year(p0, percent, aug, p):

    nb = 0
    
    while p0 < p:
        p0 += p0 * percent / 100 + aug
        nb += 1
        
    return nb

assert nb_year(1500, 5, 100, 5000) == 15
assert nb_year(1500000, 2.5, 10000, 2000000) == 10
assert nb_year(1500000, 0.25, 1000, 2000000) == 94
```

```python
# https://www.codewars.com/kata/59437bd7d8c9438fb5000004/train/python

import re

def alphabet_war(battlefield):
    
    if '#' not in battlefield:
        return battlefield.replace('[','').replace(']','')
    
    # Here (?=...) is a lookahead assertion
    matches = re.findall(r'(?=(?:^|\])([a-z#]*\[[a-z]*\][a-z#]*)(?:\[|$))', battlefield)
    survived = [re.findall(r'\[([a-z]*)\]', m)[0] for m in matches if m.count('#') < 2]
    return ''.join(survived)

assert alphabet_war('[a]#[b]#[c]') == 'ac'
assert alphabet_war('[a]#b#[c][d]') == 'd'
assert alphabet_war('[a][b][c]') == 'abc'
assert alphabet_war('##a[a]b[c]#') =='c'
assert alphabet_war('abde[fgh]ijk') == 'abdefghijk'
assert alphabet_war('ab#de[fgh]ijk') == 'fgh'
assert alphabet_war('ab#de[fgh]ij#k') == ''
assert alphabet_war('##abde[fgh]ijk') == ''
assert alphabet_war('##abde[fgh]') == ''
assert alphabet_war('##abcde[fgh]') == ''
assert alphabet_war('abcde[fgh]') == 'abcdefgh'
assert alphabet_war('##abde[fgh]ijk[mn]op') == 'mn'
assert alphabet_war('#abde[fgh]i#jk[mn]op') == 'mn'
assert alphabet_war('[ab]adfd[dd]##[abe]dedf[ijk]d#d[h]#') == 'abijk'
```

```python
# https://www.codewars.com/kata/5266876b8f4bf2da9b000362/train/python

def likes(names):

    if len(names) == 0:
        return 'no one likes this'
    
    if len(names) == 1:
        return names[0] + ' likes this'
    
    if len(names) == 2:
        return names[0] + ' and ' + names[1] + ' like this'
    
    if len(names) == 3:
        return ', '.join(names[0:2]) + ' and ' + names[2] + ' like this'
    
    if len(names) >= 4:
        return ', '.join(names[0:2]) + ' and ' + str(len(names) - 2) + ' others like this'

assert likes([]) == 'no one likes this'
assert likes(['Peter']) == 'Peter likes this'
assert likes(['Jacob', 'Alex']) == 'Jacob and Alex like this'
assert likes(['Max', 'John', 'Mark']) == 'Max, John and Mark like this'
assert likes(['Alex', 'Jacob', 'Mark', 'Max']) == 'Alex, Jacob and 2 others like this'
```
