```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: 4*(x**3) - 96*(x**2) + 576*x
x = np.arange(-2,20,0.1)
y = [f(x) for x in x]

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('volume')

plt.axvline(x=12, color='y')
plt.axhline(y=f(12), color='y')
plt.plot(12, f(12), 'ro', color='y')
plt.text(12, f(12) - 500, '({},{})'.format(12, f(12)))

plt.axvline(x=4, color='r')
plt.axhline(y=f(4), color='r')
plt.plot(4, f(4), 'ro', color='r')
plt.text(4, f(4) + 300, '({},{})'.format(4, f(4)))
```

```python
f2 = lambda x: (x-4)*(x-12)
y2 = [f2(x) for x in x]

plt.plot(x, y2)

plt.axvline(x=12, color='y')
plt.axhline(y=f2(12), color='y')
plt.plot(12, f2(12), 'ro', color='y')

plt.axvline(x=4, color='r')
plt.axhline(y=f2(4), color='r')
plt.plot(4, f2(4), 'ro', color='r')
```

```python
from scipy import optimize
f1 = lambda x: 4*(x**3) - 96*(x**2) + 576*x
f2 = lambda x: - (4*(x**3) - 96*(x**2) + 576*x)
r1 = optimize.minimize_scalar(f1, bounds=(-2, 20), method='bounded')
r2 = optimize.minimize_scalar(f2, bounds=(-2, 20), method='bounded')
print(r1.x)
print(r2.x)
```

```python
from scipy.optimize import fsolve
f = lambda x: (x-4)*(x-12)
r = fsolve(f, [0, 7])
print(r)
```
