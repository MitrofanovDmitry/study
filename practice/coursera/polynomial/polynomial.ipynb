```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as pp
%matplotlib inline
```

```python
x = np.arange(1., 15.1, 0.1)
y = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
def polynomial(x):
    n = len(x)
    A = [[x[i] ** j for j in range(n)] for i in range(n)]
    x, A = np.array(x), np.array(A)
    b = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
    return A, b

A1, b1 = polynomial([1, 15])
A2, b2 = polynomial([1, 8, 15])
A3, b3 = polynomial([1, 4, 10, 15])

w1 = linalg.solve(A1, b1)
w2 = linalg.solve(A2, b2)
w3 = linalg.solve(A3, b3)

y1 = w1[0] + w1[1] * x
y2 = w2[0] + w2[1] * x + w2[2] * (x ** 2)
y3 = w3[0] + w3[1] * x + w3[2] * (x ** 2) + w3[3] * (x ** 3)
```

```python
pp.plot(x, y,  label='y')
pp.plot(x, y1, label='y1')
pp.plot(x, y2, label='y2')
pp.plot(x, y3, label='y3')
pp.legend();
```
