```python
import numpy as np
import matplotlib.pyplot as pp
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

```python
x = np.array([1,2,3,4,5,6])
X = np.array([[1],[2],[3],[4],[5],[6]])
y = np.array([11,12,13,14,15,16])
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
print(X_train.flatten(), X_test.flatten())
print(y_train, y_test)
```

```python
x = np.arange(0,1000,1) + np.random.normal(0,1,1000)
y = np.arange(0,10000,10) + np.random.normal(0,1000,1000)
X = x.reshape(1000,1)
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
```

```python
lr = Ridge(alpha=0)
reg = lr.fit(X=X_train, y=y_train)
reg.score(X, y)
```

```python
reg.coef_, reg.intercept_
```

```python
reg.predict(np.array([[5]]))
```

```python
fig, ax = pp.subplots()
ax.scatter(X_train.flatten(), y_train)
ax.scatter(X_test.flatten(), y_test)
ax.scatter(X_test.flatten(), reg.predict(X_test))
ax.scatter(X_train.flatten(), reg.predict(X_train))
pp.show()
```
