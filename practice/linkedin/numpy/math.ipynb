```python
pp.plot(x, np.sin(x))
pp.plot(x, np.cos(x))
pp.plot(x, np.log(1 + x))
```

```python
pp.plot(x, np.sin(x), label='sinx')
pp.plot(x, np.cos(x), label='cosx')
pp.plot(x, np.log(1 + x), label='log(1+x)')
pp.legend()
```
